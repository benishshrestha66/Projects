```python
import pandas as pd
import nepali_datetime as nd

df = pd.read_excel("sabina working.xlsx")

def bs_to_ad(bs_date):
    try:
        if isinstance(bs_date, str):
            year, month, day = map(int, bs_date.split("-"))
        else:  # already a datetime/Timestamp object
            year, month, day = bs_date.year, bs_date.month, bs_date.day
        return nd.date(year, month, day).to_datetime_date()
    except Exception as e:
        print(f"FAILED: {bs_date} -> {e}")
        return None

df["ad_date"] = df["Date"].apply(bs_to_ad)

# Format as month/day/year, e.g. 4/1/2026
df["ad_date"] = df["ad_date"].apply(lambda d: d.strftime("%m/%d/%Y") if pd.notna(d) else None)

print(df)
print("Failed rows:", df["ad_date"].isna().sum())
```
